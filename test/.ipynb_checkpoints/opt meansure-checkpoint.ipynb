# torch

```python
import torch
@torch.no_grad()
class water_torch():
    def __init__(self, map_data):
        self.next_step = torch.tensor([
            [1,0],
            [0,1],
            [0, -1],
            [-1, 0],

            [1,1],
            [-1, -1],
            [-1, 1],
            [1, -1],
        ]).cuda()
        self.init(map_data)
    
    def init(self, map_data):
        map_data = torch.from_numpy(map_data)
        self.map_data = map_data.cuda()
        self.is_zeros_map = (torch.sum(map_data) == 0).cuda()
       
        self.M, self.N = map_data.shape
        self.label_map = torch.zeros_like(map_data).cuda()
        self.total = 0
        self.max = torch.tensor([-1, 0]).cuda() #标记值与联通区域id
        
        
    def dfs(self, coord, count):
        x,y = coord
        assert count != 0, 'count should be positive'
        self.total += 1
        self.label_map[x][y] = count
        for i in range(8):
            next_coord = coord + self.next_step[i]
            next_x, next_y = next_coord[0], next_coord[1] 

            if next_x <0 or next_x >= self.M or next_y >= self.N or next_y <0 :
                continue
            #print(next_coord)
            if self.label_map[next_x][next_y] == 0 and self.map_data[next_x][next_y]:
                #total += 1
                self.dfs(next_coord, count)
        return
    
    def get_max_erea(self):
        if self.is_zeros_map:
            return [0,0,0,0]
        point_id = self.max[1]
        x,y = torch.where( self.label_map == point_id )
        x = x/self.M
        y = y/self.N
        return [ torch.min(x), torch.min(y), torch.max(x), torch.max(y) ]

    def forward(self):
        count = 0
        if self.is_zeros_map:
            return count
        
        for i in range(self.M):
            for j in range(self.N):
                if self.map_data[i][j]:
                    if self.label_map[i][j] == 0:
                        count += 1
                        self.total = 0
                        self.dfs(torch.tensor([i,j]).cuda(), count)
                        if self.max[0] < self.total:
                            self.max[0] = self.total
                            self.max[1] = count
        self.total = count
        return count
                    #print(label_map)
    def __call__(self):
        pass
```

# numpy 

```python
class water():
    def __init__(self, map_data):
        self.next_step = np.array([
            [1,0],
            [0,1],
            [0, -1],
            [-1, 0],

            [1,1],
            [-1, -1],
            [-1, 1],
            [1, -1],
        ])
        self.init(map_data)
    
    def init(self, map_data):
        self.map_data = map_data
        self.is_zeros_map = np.sum(map_data) == 0
       
        self.M, self.N = map_data.shape
        self.label_map = np.zeros_like(map_data)
        self.total = 0
        self.max = [-1, 0] #标记值与联通区域id
        
    def dfs(self, coord, count):
        x,y = coord
        assert count != 0, 'count should be positive'
        self.total += 1
        self.label_map[x][y] = count
        for i in range(8):
            next_coord = coord + self.next_step[i]
            next_x, next_y = next_coord[0], next_coord[1] 

            if next_x <0 or next_x >= self.M or next_y >= self.N or next_y <0 :
                continue
            #print(next_coord)
            if self.label_map[next_x][next_y] == 0 and self.map_data[next_x][next_y]:
                #total += 1
                self.dfs(next_coord, count)
        return
    
    def get_max_erea(self):
        if self.is_zeros_map:
            return [0,0,0,0]
        point_id = self.max[1]
        x,y = np.where( self.label_map == point_id )
        x = x/self.M
        y = y/self.N
        return [ np.min(x), np.min(y), np.max(x), np.max(y) ]

    def forward(self):
        if self.is_zeros_map:
            return 0
        
        count = 0
        for i in range(self.M):
            for j in range(self.N):
                if self.map_data[i][j] and self.label_map[i][j] == 0:
                    count += 1
                    self.total = 0
                    self.dfs(np.array([i,j]), count)
                    if self.max[0] < self.total:
                        self.max[0] = self.total
                        self.max[1] = count
        self.total = count
        return count
                    #print(label_map)
    def __call__(self):
        pass
```
